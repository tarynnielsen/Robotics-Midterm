```python
import csv
import matplotlib.pyplot as plt
```

```python
def plot_csv(file_path):
    x_values = []
    y_values = []

    with open(file_path, 'r') as csvfile:
        # Create a CSV reader object
        csv_reader = csv.reader(csvfile,delimiter=' ')

        for row in csv_reader:
            # Orbslam stores a timestamp in col 0, then x in 1 and y in 2
            x_values.append(float(row[1]))
            y_values.append(float(row[2]))

    # Plot the data
    plt.plot(x_values, y_values, marker='o', linestyle='-', color='b')
    plt.title(file_path[:-4])
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.grid(True)
    plt.show()
```

```python
# KeyFrameTrajectory
csv_file_path = 'KeyFrameTrajectory.txt'
plot_csv(csv_file_path)
```

```python
# CameraTrajectory
csv_file_path = 'CameraTrajectory.txt'
plot_csv(csv_file_path)
```
